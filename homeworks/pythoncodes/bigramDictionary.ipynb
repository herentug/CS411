```python
dic = {}

alp="ABCDEFGHIJKLMNOPQRSTUVWXYZ .!,?"
```

```python
len(alp)
```

```python
val=0
for i in range(0,31):
    for j in range(0,31):
        bigram=""
        bigram+= alp[i]
        bigram+= alp[j]
        
        dic[bigram]=val
        val+=1
```

```python
print(dic)
```

```python
inv_dic = {v: k for k, v in dic.items()}
```

```python
inv_dic
```

```python

new_dict = dict([(value, key) for key, value in inv_dic.items()]) 
```
